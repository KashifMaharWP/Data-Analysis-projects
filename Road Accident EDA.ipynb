## Month-Year have Object Data Type so We will convert it into DateTime

```python
Pak_data=data.iloc[1:22]
ISL_data=data.iloc[23:44]
PNJ_data=data.iloc[45:66]
SND_data=data.iloc[67:88]
KPK_data=data.iloc[89:110]
BL_data=data.iloc[111:132]
```

```python
var=[PNJ_data['Total Number of Accident'].sum(),
     SND_data['Total Number of Accident'].sum(),
     KPK_data['Total Number of Accident'].sum(),
     BL_data['Total Number of Accident'].sum()
    ]
```

```python
plt.pie(var,labels=['Punjab','Sindh','KPK','Balochistan'],autopct='%1.1f%%',startangle=90)
plt.title('Number of Accident in Pakistan 2012-22')
plt.axis('equal')
plt.show()
```

```python
Pro21=[PNJ_data[PNJ_data['Month - Year'].str.contains('2021')]['Total Number of Accident'].sum(),
SND_data[SND_data['Month - Year'].str.contains('2021')]['Total Number of Accident'].sum(),
KPK_data[KPK_data['Month - Year'].str.contains('2021')]['Total Number of Accident'].sum(),
BL_data[BL_data['Month - Year'].str.contains('2021')]['Total Number of Accident'].sum()
]
```

```python
plt.pie(Pro21,labels=['Punjab','Sindh','KPK','Balochistan'],autopct='%1.1f%%')
plt.title('Accident Record Province Wise 2021')
plt.axis('equal')
plt.show()
```

```python
Pro12=[PNJ_data[PNJ_data['Month - Year'].str.contains('2012-2013')]['Total Number of Accident'].sum(),
SND_data[SND_data['Month - Year'].str.contains('2012-2013')]['Total Number of Accident'].sum(),
KPK_data[KPK_data['Month - Year'].str.contains('2012-2013')]['Total Number of Accident'].sum(),
BL_data[BL_data['Month - Year'].str.contains('2012-2013')]['Total Number of Accident'].sum()
]
```

```python
plt.pie(Pro12,labels=['Punjab','Sindh','KPK','Balochistan'],autopct='%1.1f%%')
plt.title('Accident Record Province Wise 2012')
plt.axis('equal')
plt.show()
```
